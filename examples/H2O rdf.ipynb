```python


def get_fcc_unit_len(rho, mass):
    
    Na = 6.02214086 * 10**23 

    N = (rho * 10**6 / mass) * Na  # number of molecules in 1m^3 of water 

    rho = N / (10 ** 30) # number density in 1 A^3

    N_mol = 4 # FCC unit cell has 4 atoms   
    L = (N_mol / rho) ** (1/3)
    
    return L 
```
